```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
easy = "0.80667 (0.09018) 0.367 (0.090)	0.290 (0.010)"
medium = "0.81000 (0.11000)	0.47000 (0.04583)	0.36000 (0.02000)"
hard = "0.77000 (0.06557)	0.43667 (0.02082)	0.37000 (0.06557)"
```

```python
easy2 = "86.19846 (9.44357)	37.402 (10.275)	28.388 (1.829)"
medium2 = "94.81751 (8.95366)	49.81355 (4.91080)	35.82628 (2.79738)"
hard2 = "92.86667 (16.00706)	45.85570 (2.56063)	37.74727 (8.39216)"
```

```python
num = lambda s: [float(s_.replace('(', '').replace(')', '')) for s_ in s.replace('\t', ' ').split(' ')]
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Create a subset of an existing colormap
cmap = plt.get_cmap('Blues')  # Choose a built-in colormap
subset_cmap = LinearSegmentedColormap.from_list(
    "subset_viridis", cmap(np.linspace(0, 0.8, 256))  # Select range 0.2 to 0.8
)
```

```python
fig, axs = plt.subplots(figsize=(8, 3), ncols=2)
axidx = 0
lable = ['easy', 'med.', 'hard']
titles = ['Objective Proportion', 'Average Rewards']
for p in [[easy, medium, hard], [easy2, medium2, hard2]]:
    lidx = 0
    means = []
    for pidx, p_ in enumerate(p):
        vals = num(p_)
        mean = [vals[i] for i in [0, 2, 4]]
        std = [vals[i] for i in [1, 3, 5]]
        means.append(mean)
        for midx, m_ in enumerate(mean):
            # f"{mean_values[i, j]:.2f}  {std_values[i, j]:.2f}"
            axs[axidx].text(midx, pidx, f'{round(m_, 2)}\n {round(std[midx], 2)}', ha='center', va='center', color='k') #, fontsize=8)
    cbar = axs[axidx].imshow(means, cmap=subset_cmap) #, marker='o', linestyle='-',)
    lidx += 1
    cbar = plt.colorbar(cbar, ax=axs[axidx])
    axs[axidx].set_xticks(range(3), lable, rotation=30)
    axs[axidx].set_yticks(range(3), lable, rotation=30)
    axs[axidx].set_title(titles[axidx])
    axs[axidx].set_xlabel('Inference Simulation Difficulty')
    axidx+=1
# plt.legend(ncol=3, bbox_to_anchor=(1.05, 1.25), title='Trained Simulation Difficulty')

axs[0].set_ylabel('Trained Simulation Difficulty')
plt.suptitle('DDQN')
# plt.tight_layout()
plt.savefig('ddqn_heat.png',dpi = 300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
mean = f.iloc[:3000, 1:4].to_numpy().mean(axis=-1)
std =  f.iloc[:3000, 1:4].to_numpy().std(axis=-1)
ax.plot(range(3000), mean, color='red', label='DDQN')
ax.fill_between(range(3000), mean-std, mean+std, color='red', alpha=.2)

mean = f.iloc[:3000, 4:7].to_numpy().mean(axis=-1)
std =  f.iloc[:3000, 4:7].to_numpy().std(axis=-1)
ax.plot(range(3000), mean, color='blue', label='PPO')
ax.fill_between(range(3000), mean-std, mean+std, color='blue', alpha=.2)

ax.set_xlabel('Steps')
ax.set_ylabel('Objective Proportion (logscale)')

ax.set_yscale('log')
ax.legend()
plt.savefig('fig1.png', dpi=300, bbox_inches='tight')
```
